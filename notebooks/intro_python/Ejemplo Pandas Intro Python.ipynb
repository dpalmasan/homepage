````python
import pandas as pd


csv_url = "https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv"

dataset = pd.read_csv(csv_url)

# Siempre es buena idea ver cómo son los datos
dataset.head()
````

````python
# Información del dataset (Uso en memoria, tipos, etc)
dataset.info()
````

````python
# Estadística descriptiva
dataset.describe()
````

````python
# También se puede acceder a información de algunas columnas del dataset
dataset[["mpg", "hp"]].head()
````

````python
# Podemos también contar valores:
dataset["cyl"].value_counts()
````

````python
# También podemos calcular los porcentajes:
porcentajes = (dataset["cyl"].value_counts()/dataset.shape[0]).reset_index(name="count")
porcentajes
````

````python
# Podemos también calcular el promedio de una columna
dataset["hp"].mean()
````

````python
# Podemos elegir sub-conjuntos en base a algun criterio
dataset[dataset["hp"] > dataset["hp"].mean()].head()
````

````python
# Podemos ordenar en base a alguna columna o columnas
dataset.sort_values(by=["wt", "qsec"]).head()
````

````python
# También en forma descendente
dataset.sort_values(by=["wt", "qsec"], ascending=False).head()
````
